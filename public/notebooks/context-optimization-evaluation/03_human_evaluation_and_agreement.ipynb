## 2. Cohen's Kappa: The Mathematics

$$\kappa = \frac{p_o - p_e}{1 - p_e}$$

**Worked example:** 100 samples, two annotators:

| | B: Good | B: Bad |
|--|---------|--------|
| A: Good | 40 | 10 |
| A: Bad | 15 | 35 |

$p_o = (40+35)/100 = 0.75$, $p_e = 0.275 + 0.225 = 0.5$

$\kappa = (0.75 - 0.5)/(1 - 0.5) = 0.5$ (moderate agreement)

```python
#@title 🎧 Listen: Kappa Impl
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/03_06_kappa_impl.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
def cohens_kappa(labels_a, labels_b):
    assert len(labels_a) == len(labels_b)
    n = len(labels_a)
    agreements = sum(1 for a, b in zip(labels_a, labels_b) if a == b)
    p_o = agreements / n

    all_labels = sorted(set(labels_a) | set(labels_b))
    p_e = sum((labels_a.count(l)/n) * (labels_b.count(l)/n) for l in all_labels)

    kappa = (p_o - p_e) / (1 - p_e) if p_e < 1 else 1.0

    if kappa > 0.8: interp = "Almost perfect"
    elif kappa > 0.6: interp = "Substantial"
    elif kappa > 0.4: interp = "Moderate"
    elif kappa > 0.2: interp = "Fair"
    else: interp = "Slight/poor"

    return {"kappa": kappa, "p_o": p_o, "p_e": p_e,
            "interpretation": interp, "n": n, "agreed": agreements}

# Reproduce worked example
ann_a = ["Good"]*40 + ["Good"]*10 + ["Bad"]*15 + ["Bad"]*35
ann_b = ["Good"]*40 + ["Bad"]*10 + ["Good"]*15 + ["Bad"]*35
result = cohens_kappa(ann_a, ann_b)
print("=== Cohen's Kappa ===")
for k, v in result.items():
    print(f"  {k}: {v:.3f}" if isinstance(v, float) else f"  {k}: {v}")
```

## 3. Likert Scale Evaluation

```python
#@title 🎧 Listen: Likert Impl
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/03_08_likert_impl.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
class LikertEvaluator:
    DIMS = ["relevance", "completeness", "coherence", "faithfulness"]

    def __init__(self):
        self.evaluations = []

    def add_eval(self, sample_id, annotator, ratings):
        self.evaluations.append({"sample_id": sample_id, "annotator": annotator, **ratings})

    def averages(self):
        df = pd.DataFrame(self.evaluations)
        return df.groupby("sample_id")[self.DIMS].mean().round(2)

    def agreement(self):
        df = pd.DataFrame(self.evaluations)
        results = {}
        annotators = df["annotator"].unique()
        if len(annotators) < 2: return results
        for dim in self.DIMS:
            a1 = df[df["annotator"]==annotators[0]]
            a2 = df[df["annotator"]==annotators[1]]
            common = set(a1["sample_id"]) & set(a2["sample_id"])
            la = ["High" if a1[a1["sample_id"]==s][dim].values[0]>=4 else "Low" for s in sorted(common)]
            lb = ["High" if a2[a2["sample_id"]==s][dim].values[0]>=4 else "Low" for s in sorted(common)]
            results[dim] = cohens_kappa(la, lb)
        return results

np.random.seed(42)
ev = LikertEvaluator()
for sid in range(20):
    base = {d: int(np.random.choice([3,4,5], p=[0.2,0.4,0.4])) for d in ev.DIMS}
    ev.add_eval(sid, "Ann_A", base)
    noisy = {d: max(1, min(5, v + int(np.random.choice([-1,0,0,0,1])))) for d, v in base.items()}
    ev.add_eval(sid, "Ann_B", noisy)

avgs = ev.averages()
print("=== Average Ratings ===")
print(avgs.head(10))
print(f"\nOverall: {avgs.mean().round(3).to_dict()}")
```

## 4. Pairwise Comparison

```python
#@title 🎧 Listen: Pairwise Impl
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/03_11_pairwise_impl.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
class PairwiseComparator:
    def __init__(self):
        self.comparisons = []

    def add(self, sample_id, annotator, winner):
        self.comparisons.append({"sample_id": sample_id, "annotator": annotator, "winner": winner})

    def winrate(self):
        df = pd.DataFrame(self.comparisons)
        c = df["winner"].value_counts()
        t = len(df)
        return {"A": c.get("A",0), "B": c.get("B",0), "tie": c.get("tie",0),
                "A_rate": c.get("A",0)/t, "B_rate": c.get("B",0)/t}

    def agreement(self):
        df = pd.DataFrame(self.comparisons)
        anns = df["annotator"].unique()
        if len(anns) < 2: return {}
        a1 = df[df["annotator"]==anns[0]].set_index("sample_id")["winner"]
        a2 = df[df["annotator"]==anns[1]].set_index("sample_id")["winner"]
        common = sorted(set(a1.index) & set(a2.index))
        return cohens_kappa([a1[s] for s in common], [a2[s] for s in common])

np.random.seed(42)
comp = PairwiseComparator()
for sid in range(30):
    for ann in ["Ann_1", "Ann_2"]:
        r = np.random.random()
        w = "A" if r < 0.25 else "B" if r < 0.85 else "tie"
        comp.add(sid, ann, w)

wr = comp.winrate()
print("=== Pairwise Results ===")
print(f"  System A: {wr['A']} ({wr['A_rate']:.1%})")
print(f"  System B: {wr['B']} ({wr['B_rate']:.1%})")
print(f"  Ties: {wr['tie']}")
agr = comp.agreement()
print(f"  Agreement: kappa={agr['kappa']:.3f} ({agr['interpretation']})")
```
