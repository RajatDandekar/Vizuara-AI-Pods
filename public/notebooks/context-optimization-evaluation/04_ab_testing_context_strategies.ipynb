## 2. The Mathematics: Two-Proportion Z-Test

$$z = \frac{\hat{p}_B - \hat{p}_A}{\sqrt{\hat{p}(1-\hat{p})\left(\frac{1}{n_A} + \frac{1}{n_B}\right)}}$$

**Worked example:** A: 300/500 = 60%, B: 335/500 = 67%, pooled = 0.635

$z = 0.07 / \sqrt{0.635 \times 0.365 \times 0.004} = 0.07 / 0.0305 = 2.30$

p-value = 0.021 < 0.05 => significant!

```python
#@title 🎧 Listen: Ztest Function
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/04_06_ztest_function.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
def two_proportion_ztest(sa, na, sb, nb):
    pa, pb = sa/na, sb/nb
    pp = (sa + sb) / (na + nb)
    se = math.sqrt(pp * (1-pp) * (1/na + 1/nb))
    z = (pb - pa) / se if se > 0 else 0
    pval = 2 * (1 - stats.norm.cdf(abs(z)))
    se_diff = math.sqrt(pa*(1-pa)/na + pb*(1-pb)/nb)
    ci = ((pb-pa) - 1.96*se_diff, (pb-pa) + 1.96*se_diff)
    return {"p_a": pa, "p_b": pb, "lift": pb-pa, "z": z, "p_value": pval,
            "significant": pval < 0.05, "ci_95": ci}

result = two_proportion_ztest(300, 500, 335, 500)
print("=== A/B Test: Few-Shot Examples ===")
print(f"Control:   {result['p_a']:.1%}")
print(f"Treatment: {result['p_b']:.1%}")
print(f"Lift:      +{result['lift']:.1%}")
print(f"Z-score:   {result['z']:.2f}")
print(f"P-value:   {result['p_value']:.4f}")
print(f"95% CI:    ({result['ci_95'][0]:+.3f}, {result['ci_95'][1]:+.3f})")
print(f"Significant: {result['significant']}")
```

## 3. Sample Size Calculator

Before running an experiment, calculate how many samples you need.

```python
#@title 🎧 Listen: Sample Size Function
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/04_08_sample_size_function.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
def required_sample_size(baseline, mde, alpha=0.05, power=0.80):
    p1, p2 = baseline, baseline + mde
    p_avg = (p1 + p2) / 2
    z_a = stats.norm.ppf(1 - alpha/2)
    z_b = stats.norm.ppf(power)
    n = ((z_a * math.sqrt(2*p_avg*(1-p_avg)) + z_b * math.sqrt(p1*(1-p1)+p2*(1-p2))) / (p2-p1))**2
    return math.ceil(n)

print("=== Sample Size Calculator ===")
for name, base, mde in [("Small +3%", 0.60, 0.03), ("Medium +5%", 0.60, 0.05), ("Large +10%", 0.60, 0.10)]:
    n = required_sample_size(base, mde)
    print(f"  {name:<15} n={n:,} per group ({2*n:,} total)")
```

## 4. Experiment Simulator

```python
#@title 🎧 Listen: Simulator Class
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/04_11_simulator_class.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
class ABTestSimulator:
    def __init__(self, true_pa, true_pb, seed=42):
        self.pa, self.pb = true_pa, true_pb
        self.rng = np.random.RandomState(seed)

    def run(self, n):
        ca = self.rng.binomial(1, self.pa, n)
        cb = self.rng.binomial(1, self.pb, n)
        return two_proportion_ztest(ca.sum(), n, cb.sum(), n)

    def power_analysis(self, sizes, n_sims=500):
        return [sum(1 for _ in range(n_sims)
                    if ABTestSimulator(self.pa, self.pb, seed=_).run(n)['significant']
                ) / n_sims for n in sizes]

sim = ABTestSimulator(0.60, 0.67)
r = sim.run(500)
print(f"True effect: +7%")
print(f"Observed: {r['p_a']:.1%} vs {r['p_b']:.1%}, lift={r['lift']:+.1%}")
print(f"z={r['z']:.2f}, p={r['p_value']:.4f}, significant={r['significant']}")
```

```python
#@title 🎧 Listen: Power Plot
from IPython.display import Audio, display
import os as _os
_f = "/content/narration/04_12_power_plot.mp3"
if _os.path.exists(_f):
    display(Audio(_f))
else:
    print("Run the first cell to download narration audio.")
```

```python
sizes = [50, 100, 200, 300, 500, 750, 1000]
powers = sim.power_analysis(sizes, n_sims=300)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(sizes, powers, 'bo-', linewidth=2, markersize=8)
ax.axhline(y=0.80, color='red', linestyle='--', label='80% power')
ax.set_xlabel('Samples Per Group')
ax.set_ylabel('Power')
ax.set_title('Power Analysis: +7% Effect')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('power_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```
