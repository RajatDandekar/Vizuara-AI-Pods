```python
# ============================================================
#  Scoring Engine
# ============================================================

def score_exam(answers, answer_key, domains):
    """Score the practice exam and display detailed results."""

    # Validate answers
    unanswered = []
    invalid = []
    for q_id in answer_key:
        val = answers.get(q_id, "").strip().upper()
        if val == "":
            unanswered.append(q_id)
        elif val not in ("A", "B", "C", "D"):
            invalid.append((q_id, answers[q_id]))

    if invalid:
        print("INVALID ANSWERS DETECTED:")
        for q_id, val in invalid:
            print(f"  {q_id}: '{val}' — must be 'A', 'B', 'C', or 'D'")
        print()

    if unanswered:
        print(f"WARNING: {len(unanswered)} unanswered question(s): {', '.join(unanswered)}")
        print("Unanswered questions are scored as incorrect. No penalty for guessing!\n")

    # Calculate per-domain results
    print("=" * 65)
    print("  CLAUDE CERTIFIED ARCHITECT — PRACTICE EXAM RESULTS")
    print("=" * 65)
    print()

    domain_scores = {}
    total_correct = 0
    total_questions = 0

    for d_id, d_info in domains.items():
        correct = 0
        total = len(d_info["questions"])
        missed = []

        for q_id in d_info["questions"]:
            student_answer = answers.get(q_id, "").strip().upper()
            if student_answer == answer_key[q_id]:
                correct += 1
            else:
                missed.append(q_id)

        pct = (correct / total) * 100 if total > 0 else 0
        domain_scores[d_id] = {
            "correct": correct,
            "total": total,
            "pct": pct,
            "missed": missed,
            "weight": d_info["weight"],
            "name": d_info["name"],
        }

        total_correct += correct
        total_questions += total

    # Calculate weighted scaled score (100–1000 range)
    weighted_sum = 0
    for d_id, ds in domain_scores.items():
        weighted_sum += (ds["pct"] / 100.0) * ds["weight"]

    # Scale: 0% -> 100, 100% -> 1000
    scaled_score = int(100 + weighted_sum * 900)
    passed = scaled_score >= 720

    # Display domain breakdown
    print("  DOMAIN BREAKDOWN")
    print("-" * 65)

    for d_id in sorted(domain_scores.keys()):
        ds = domain_scores[d_id]
        bar_filled = int(ds["pct"] / 5)   # 20 chars = 100%
        bar_empty = 20 - bar_filled
        bar = "#" * bar_filled + "." * bar_empty

        if ds["pct"] >= 80:
            indicator = "[STRONG]"
        elif ds["pct"] >= 60:
            indicator = "[  OK  ]"
        else:
            indicator = "[ WEAK ]"

        weight_str = f"(weight: {ds['weight']:.0%})"
        print(f"  {d_id}: {ds['name']}")
        print(f"       {ds['correct']}/{ds['total']} correct ({ds['pct']:.0f}%)  "
              f"[{bar}]  {indicator}  {weight_str}")

        if ds["missed"]:
            print(f"       Missed: {', '.join(ds['missed'])}")
        print()

    # Overall results
    raw_pct = (total_correct / total_questions) * 100 if total_questions > 0 else 0

    print("-" * 65)
    print(f"  RAW SCORE:     {total_correct} / {total_questions} ({raw_pct:.1f}%)")
    print(f"  SCALED SCORE:  {scaled_score} / 1000")
    print(f"  PASS MARK:     720 / 1000")
    print()

    if passed:
        print("  ===================================================")
        print("  ||                                               ||")
        print("  ||       RESULT:  P A S S                        ||")
        print("  ||                                               ||")
        print("  ||   You are ready for the certification exam!   ||")
        print("  ||                                               ||")
        print("  ===================================================")
    else:
        print("  ===================================================")
        print("  ||                                               ||")
        print("  ||       RESULT:  NOT YET PASSING                ||")
        print("  ||                                               ||")
        print("  ||   Review the weak domains below and retake.   ||")
        print("  ||                                               ||")
        print("  ===================================================")

        # Study recommendations
        print()
        print("  STUDY RECOMMENDATIONS:")
        weak_domains = [(d_id, ds) for d_id, ds in domain_scores.items() if ds["pct"] < 70]
        weak_domains.sort(key=lambda x: x[1]["pct"])

        if weak_domains:
            for d_id, ds in weak_domains:
                print(f"  -> {d_id}: {ds['name']} ({ds['pct']:.0f}%) — "
                      f"Review this domain. Missed: {', '.join(ds['missed'])}")
        else:
            print("  -> No critically weak domains, but aim for 80%+ in each.")

    print()
    print("=" * 65)
    print()

    return scaled_score, passed


# Run the scoring
scaled_score, passed = score_exam(answers, ANSWER_KEY, DOMAINS)
```
